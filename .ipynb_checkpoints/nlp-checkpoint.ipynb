```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv("travel_rajasthan.csv")
df.head()
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

document1 = []
stopwords = set(stopwords.words('english'))

for tweet in df["Tweet"]:
    word_token=word_tokenize(tweet)
    
    stre=''
    for w in word_token:
        if w not in stopwords:
            stre=stre+w+''
    document1.append(stre)
  

    
df.drop("Unnamed: 3",axis=1,inplace=True)
df.drop("Unnamed: 2",axis=1,inplace=True)
df["remove_stop_words"]=document1

df.head()
```

```python
import nltk

def process_content():
	try:
		for tweet in df["Tweet"]:
			words = word_tokenize(tweet)
			tagged = nltk.pos_tag(words)
			print(tagged)
	except Exception as e:
		print(str(e))
process_content()


        
```

```python
#sentiment analysis
from textblob import TextBlob

document2 = []
document3 = []

for words in df["Tweet"]:
    analysis = TextBlob(words)
    #print(analysis.sentiment)
    document2.append(analysis.sentiment[0])
#    if analysis.sentiment[0] > 0:
#        df["Sentiment"] = "Positive"
#    else:
#        df["Sentiment"] = "Negative"

df["polarity"]=document2

for i in df["polarity"]:
    print(i)
    if i < 0.0:
        document3.append("negative")
    else:
        document3.append("positive")

df["sentiment"]=document3
df.head(10)    
```

```python
df["polarity"].plot(kind="bar")
```

```python
slices = [988,106]
activities = ['positive','negative']
cols = ['c','m']
plt.pie(slices,labels=activities,colors=cols,startangle=90,shadow=True,explode=(0,0.1),autopct='%1.1f%%')

print(slices)
plt.show()
```
